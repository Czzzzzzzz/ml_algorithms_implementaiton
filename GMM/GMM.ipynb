```python
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

# 1.Generate Toy Datasets

```python
def generate_toy_dataset(cluster_size=4, samples_per_cluster=50):
    N = cluster_size * samples_per_cluster
    y = np.zeros(N)
    x = np.random.standard_normal(size=(N, 2))
    for i in range(cluster_size):
        from_idx = i * samples_per_cluster
        to_idx = (i+1) * samples_per_cluster
        y[from_idx: to_idx] = i
        theta = 2 * np.pi * i / cluster_size
        x[from_idx: to_idx] = x[from_idx: to_idx] + (cluster_size * np.cos(theta), cluster_size * np.sin(theta))
        
    return x, y

def plot_toy_dataset(x, y):
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
        
    axes[0].scatter(x[:, 0], x[:, 1])    
    axes[1].scatter(x[:, 0], x[:, 1], c=y)

n_cluster = 4
x, y = generate_toy_dataset(4, 50)
plot_toy_dataset(x, y)    
```

# 2. GMM Model

```python
class GMM():
    '''
        Fits a Gausian Mixture model to the data.

        attrs:
            n_cluster : Number of mixtures (Int)
            e : error tolerance (Float) 
            max_iter : maximum number of updates (Int)
            means : means of Gaussian mixtures (n_cluster X D numpy array)
            variances : variance of Gaussian mixtures (n_cluster X D X D numpy array) 
            pi_k : mixture probabilities of different component ((n_cluster,) size numpy array), which 
                    represents the prior probability of the latent variable z (or i here).
    '''

    def __init__(self, n_cluster, max_iter=100, e=0.0001):
        self.n_cluster = n_cluster
        self.e = e
        self.max_iter = max_iter
        self.means = None
        self.variances = None
        self.pi_k = None

    def fit(self, x):
        '''
            Fits a GMM to x. Updates self.means, self.variances, and self.pi_k

            x: a NXD size numpy array
        '''
        assert len(x.shape) == 2, 'x can only be 2 dimensional'

        np.random.seed(42)
        N, D = x.shape

        # The main steps are as follows:
        # 1. initialize means, variances, and pi_k
        # 2. In the E-step, compute gamma_ik which is the posterior distribution of the latent variable.
        # 3. In the M-step, update mean, variances, and pi_k
        # 4. update until convergence or until you have made self.max_iter updates.
        
        # Randomly initialize self.means, self.variances, and self.pi_k
        self.means = np.array([[np.random.rand() for _ in range(D)] for _ in range(self.n_cluster)])
        self.variances = np.array([np.identity(D) for _ in range(self.n_cluster)])
        self.pi_k = np.ones(self.n_cluster) / self.n_cluster

        def update_likelihood():            
            gaussian_pdfs = [self.Gaussian_pdf(self.means[k], self.variances[k]) for k in range(self.n_cluster)]

            likelihood_ik = []
            for sample_idx in range(N):
                likelihood = np.array([self.pi_k[k] * gaussian_pdfs[k].getLikelihood(x[sample_idx]) for k in range(self.n_cluster)])
                normalization = np.sum(likelihood)
                gamma_i = likelihood / normalization
                likelihood_ik.append(gamma_i)
            return np.array(likelihood_ik)

        loss = self.compute_log_likelihood(x, self.means, self.variances, self.pi_k)
        
        n_updates = 0
        for iter in range(self.max_iter):
            n_updates += 1

            #compute gamma_ik
            gamma_ik = update_likelihood()

            #update means, variance, and pi_k
            n_k = np.sum(gamma_ik, axis=0)

            self.means = np.array([np.sum(gamma_ik[:, k].reshape(-1, 1) * x, axis=0) / n_k[k] for k in range(self.n_cluster)])
            self.variances = np.array([np.matmul(gamma_ik[:, k]*(x-self.means[k]).T, (x-self.means[k])) / n_k[k] for k in range(self.n_cluster)])
            self.pi_k = n_k / N

            loss_new = self.compute_log_likelihood(x, self.means, self.variances, self.pi_k)

            if abs(loss_new - loss) < self.e:
                break
            else:
                loss = loss_new

        return n_updates

    def sample(self, N):
        '''
        sample from the GMM model

        N is a positive integer
        return : NXD array of samples
        '''
        assert type(N) == int and N > 0, 'N should be a positive integer'
        np.random.seed(42)
        if (self.means is None):
            raise Exception('Train GMM before sampling')
        
        # pi_k tells us the probability the sample belonging to the cluster i is picked up.
        # self.means and self.variances tells us the multivariate normal distribution in each cluster.
        clusters = np.random.choice(self.n_cluster, N, p=self.pi_k)
        samples = np.array([np.random.multivariate_normal(self.means[clusters[i]], self.variances[clusters[i]]) for i in range(N)])

        return samples        

    def compute_log_likelihood(self, x, means=None, variances=None, pi_k=None):
        '''
            Return log-likelihood for the data

            x is a NXD matrix
            return : a float number which is the log-likelihood of data
        '''
        assert len(x.shape) == 2,  'x can only be 2 dimensional'
        if means is None:
            means = self.means
        if variances is None:
            variances = self.variances
        if pi_k is None:
            pi_k = self.pi_k    
            
        N, D = x.shape
        
        gaussian_pdfs = [self.Gaussian_pdf(means[k], variances[k]) for k in range(self.n_cluster)]

        # calculate log-likelihood using means, variances and pi_k                  
        log_likelihood = 0
        for i in range(N):
            loss_single = 0
            for k in range(self.n_cluster):

                gaussian = gaussian_pdfs[k]
                loss_single += pi_k[k] * gaussian.getLikelihood(x[i])

            log_likelihood += np.log(loss_single)

        log_likelihood = float(log_likelihood)

        return log_likelihood

    class Gaussian_pdf():
        '''
        Modularize the process of computing likelihood.
        '''
        def __init__(self,mean,variance):            
            self.mean = mean
            self.variance = variance
            self.c = None
            self.inv = None
            
            '''
                Input: 
                    Means: A 1 X D numpy array of the Gaussian mean
                    Variance: A D X D numpy array of the Gaussian covariance matrix
                Output: 
                    None: 
            '''            

            D = self.mean.shape[0]

            # Ensure the variance matrix is full rank.
            while np.linalg.det(variance) == 0:
                variance += 1e-3 * np.identity(D)            

            PI = 3.1415

            self.inv = np.linalg.inv(variance)
            self.c = (2*PI)**D * np.linalg.det(variance)

        def getLikelihood(self,x):
            '''
                Input: 
                    x: a 1 X D numpy array representing a sample
                Output: 
                    p: a numpy float, the likelihood sample x was generated by this Gaussian 
                    p = e^(-0.5(x-mean)*(inv(variance))*(x-mean)'/sqrt(c)) where ' is transpose and * is matrix multiplication
            '''
            # Calculate the likelihood of sample x generated by this Gaussian
            x_minus_mean = x - self.mean         
            p = np.exp(-1 / 2 * np.matmul(np.matmul(x_minus_mean, self.inv), x_minus_mean)) / np.sqrt(self.c)            
            return p
```

```python
def compute_elipse_params(variance):
    '''
        Compute elipse params for plotting from variance
    '''

    # Reference:
    # http://www.cs.cornell.edu/cv/OtherPdf/Ellipse.pdf Slide 17
    # https://stackoverflow.com/a/41821484

    variance_inv = np.linalg.inv(variance)
    a = variance_inv[0, 0]
    c = variance_inv[1, 1]
    b = variance_inv[0, 1] + variance_inv[1, 0]

    M = (variance_inv + variance_inv.T) / 2
    eig, _ = np.linalg.eig(M)
    if (np.abs(eig[0] - a) < np.abs(eig[0] - c)):
        lambda1, lambda2 = eig
    else:
        lambda2, lambda1 = eig

    angle = np.arctan(b / (a - c)) / 2
    return np.sqrt(1 / lambda1), np.sqrt(1 / lambda2), angle
```
